```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
```

```python
data=pd.read_csv('E:\Data_science\movie_predection\cost_revenue_clean.csv.csv')
data.describe()
```

```python
X = pd.DataFrame(data,columns=['production_budget_usd'])
y = pd.DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.title("Film Cost vs. Global Revenue")
plt.xlabel("Production Budget $")
plt.xlabel("Worldwide gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

regression = linear regression
Fitting X and y in regression

```python
reg= lr()
reg.fit(X,y)
```

Regression cofficient (slope m)

```python
reg.coef_
```

Regression intercept(Constant c)

```python
reg.intercept_
```

Plotting the predection line:
plt.plot(Variable X,regression.predict(variable X),making color red adn thickness of line 2)

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y,alpha=0.3)
plt.plot(X,reg.predict(X),color='red',linewidth=2)
plt.title("Film Cost vs. Global Revenue")
plt.xlabel("Production Budget $")
plt.xlabel("Worldwide gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

regression score tells us we can predict upto around 55% movies revenue accurately with this model

```python
reg.score(X,y)
```
